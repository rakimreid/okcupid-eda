```python
# Import packages 

import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Set file path to variable

file = "C:/Portfolio/Data Engineer Track/OkCupid Dataset/okcupid_profiles.csv"
```

```python
# Create dataframe

df = pd.read_csv(file, low_memory = False)

# Show all columns

pd.set_option("display.max_columns", None)

# Show all lines if needed (just remove the comment)

# pd.set_option('display.max_rows', None)
```

```python
# Rename essay1 column about careers to career 

df.rename(columns = {"essay1": "career"}, inplace = True)
df["career"] = df["career"].astype(str)

# Change column data types to numeric

df["age"] = df["age"].astype(str).astype(int)


df.info()
```
